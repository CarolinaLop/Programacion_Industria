```python
df_espanol = df.replace(np.nan,'', regex=True)
df_espanol = df_espanol[df_espanol['languages'].str.contains('es')]
df_espanol
```

```python
df_espanol.set_index('alpha_3')[['population','area']].sort_values(['population']).plot(kind='bar',rot=65, figsize=(20,10))
```

```python
def detect_anomalies(data):
  anomalies = []
  data_std = data.std()
  data_mean = data.mean()
  anomaly_cut_of = data_std * 2
  lower_limit = data_mean - anomaly_cut_of
  upper_limit = data_mean + anomaly_cut_of
  print(f'el limite superior es {upper_limit.iloc[0]} y el límite inferior es {lower_limit.iloc[0]}')

  for index, row in data.iterrows():
    outlier = row
    if outlier.iloc[0] > upper_limit.iloc[0] or outlier.iloc[0] < lower_limit.iloc[0]:
      anomalies.append(index)
  
  return anomalies

detect_anomalies(df_espanol.set_index('alpha_3')[['population']])
```

detección de Outliers, en este caso definimos como limite superior (e inferior) la media más (menos) “2 veces la desviación estándar” que muchas veces es tomada como máximos de tolerancia.
